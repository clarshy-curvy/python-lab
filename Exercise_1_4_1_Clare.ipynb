```python
#Task 1
def interest(principal, rate, time):
  interest_amount = (principal * rate * time) / 100
  return interest_amount

result = interest(100000, 5, 3)
print("Interest owed:", result)
```

```python
#Task 2
def celsius_to_fahrenheit(tc):
  fahrenheit = tc * 1.8 + 32
  return fahrenheit

print(celsius_to_fahrenheit(100))
print(celsius_to_fahrenheit(75))
print(celsius_to_fahrenheit(120))
```
